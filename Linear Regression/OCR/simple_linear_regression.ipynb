```python
import numpy as np
import pandas as pd
```

```python
class LinearRegression:
    def __init__(self):
        self.slop = 0
        self.intercept = 0  
    
    def fit(self, X, y):
        y_mean = y.mean()
        x_mean = X.mean()
        
        
        numerator_val = 0
        ddenomintaor_val = 0
        
        # m = Sum((yi - yMean)(xi - XMean)) / Sum(xi - XMean^)**2;
        
        for i in range(X.shape[0]):
            numerator_val += (X[i] - x_mean) * (y[i] - y_mean)
            ddenomintaor_val += (X[i] -x_mean)**2
            
        
        self.slop = numerator_val / ddenomintaor_val
        
        self.intercept = y_mean - (self.slop * x_mean)
        
    def predict(self, X):
        
        # y = mx + c
        
        return self.slop * X + self.intercept
```

# Generating The random Regressive values using numpy

```python
np.random.seed(42)
X = np.random.rand(50, 1) * 10  # 50 points between 0 and 10
y = 2.5 * X + np.random.randn(50, 1) * 2  # y = 2.5x + noise
```

```python
X
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, color='blue', label='Data Points')
```

```python
lgModel = LinearRegression()
```

```python
lgModel.fit(X, y)
```

```python
y_pred = lgModel.predict(X)
```

```python
plt.scatter(X, y, color = "blue")
plt.plot(X, y_pred, color = "red")
```
